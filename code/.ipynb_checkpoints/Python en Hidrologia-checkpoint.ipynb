```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
%config Completer.use_jedi = False #for auto complete code 
```

```python
rng = pd.date_range('1981-01-01','2016-12-31', freq='d')
obs=pd.read_csv('../data/Estaciones/RAIN_GAUGE.csv')
obs_daily=obs.iloc[:,3:13152].transpose()
obs_daily=obs_daily.set_index(rng)
ubicaciones = obs.iloc[:,0:2]
```

```python
ubicaciones.head()
```

```python
obs_daily.head(5)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(obs_daily.isnull(), cbar=False)
plt.show()
```

```python
filtro = pd.date_range('1990-01-01','2000-12-31', freq='d')
```

```python
obs_diarias_filtrada=obs_daily.loc[filtro]
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(obs_diarias_filtrada.isnull(), cbar=False)
plt.show()
```

```python
n=0.95*obs_diarias_filtrada.shape[0]
```

```python
col_idx=obs_diarias_filtrada.count(axis=0)>=n
```

```python
obs_diarias_filtradas_completas=obs_diarias_filtrada.loc[:,col_idx]
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(obs_diarias_filtradas_completas.isnull(), cbar=False)
plt.show()
```

```python
datos_mensuales=obs_diarias_filtradas_completas.groupby(pd.Grouper(freq='M')).agg(np.sum)
```

```python
datos_mensuales
```

# Ubicaciones en un mapa

```python
import folium
```

```python
datos_filtrados_ubicaciones=ubicaciones.loc[datos_mensuales.columns,:]
```

```python
Xc=datos_filtrados_ubicaciones.x.mean()
Yc=datos_filtrados_ubicaciones.y.mean()

mapa = folium.Map(location=[Yc,Xc],zoom_start=5)
```

```python
mapa
```

```python
k=0
for i,pos in datos_filtrados_ubicaciones.iterrows():
    popup_name='Estacion: '+str(datos_mensuales.columns[k])
    folium.CircleMarker(location=[pos.y,pos.x],popup=popup_name,radius=2).add_to(mapa)
    k=k+1
```

```python
mapa
```

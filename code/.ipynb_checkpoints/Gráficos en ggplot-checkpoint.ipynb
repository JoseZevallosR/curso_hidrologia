```r
#Liberia para gestionar fechas
library(lubridate)
#libreria para manejar bases de datos
library(dplyr)
#liberia para realizar graficos
library(ggplot2)
```

Para los ejemplos siguientes trabajaremos con los datos de una estación pluviométrica

```r
estacion=read.csv('D:/Proyectos_GitHub/curso_hidrologia/data/Estaciones/estacion9.csv')
```

Explorando las primeras 5 filas de informacion

```r
head(estacion,5)
```

Podemos observar que tenemos una columna para fechas y otra para los valores de precipitación registrados. 
Para poder realizar operaciones con fecha indexada, debemos primero transformar las fechas a un formato que R pueda procesar.

```r
estacion=estacion%>%
  mutate(fecha = as.Date(fecha, format="%m/%d/%Y"))
head(estacion,10)
```

# Primer ggplot

Nuestro primer grafico será de la serie temporal

```r
ggplot(data=estacion,aes(x=fecha,y=pp))+geom_line()
```

agregando color

```r
ggplot(data=estacion,aes(x=fecha,y=pp))+geom_line(color='blue')
```

Modificando la etiqueta de los ejes x-y. Vamos a cambiar el texto pp por precipitación y dejar en blanco la etiqueta del eje x

```r
ggplot(data=estacion,aes(x=fecha,y=pp))+geom_line(color='blue')+ylab('Precipitación (mm)') +xlab(element_blank())
```

```python
estaciones=read.csv('D:/Proyectos_GitHub/curso_hidrologia/data/Estaciones/hourly_data.csv',sep = ';')
```

```python
head(estaciones)
```

```python
names(estaciones)=c('fecha',unlist(lapply(1:(dim(estaciones)[2]-1),FUN = function(x){ paste0('Estacion-',as.character(x)) })))
```

```python
head(estaciones,10)
```

```python
estaciones = estaciones %>% mutate(fecha = as.POSIXct(fecha, format="%d/%m/%Y %H:%M") )
estaciones
```

```python
estaciones_6h= estaciones%>%
  mutate(dt_aggregate = floor_date(fecha, unit = "3 hour"))
estaciones_6h
```

```python
df=data.frame(
  DateTime=as.POSIXct(c("2030-01-01 01:00:00", "2030-01-01 01:15:00",
                        "2030-01-01 01:15:00","2030-01-01 02:15:00",
                        "2030-01-01 03:15:00","2030-01-01 03:15:00")),
  temperature=c(10,15,2,10,5,6))
df

df = df %>%
  mutate(dt_aggregate = floor_date(DateTime, unit = "hour")) %>%
  group_by(dt_aggregate) %>%
  summarise(temperature = sum(temperature, na.rm = TRUE))
```

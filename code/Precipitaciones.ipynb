```python
# https://github.com/mmhs013/pyHomogeneity

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel('../data/Estaciones/RainfallData.xlsx', sheet_name = '2MassCurve', skiprows = 4, usecols = range(6))

df['AVG'] = df.mean(axis = 1)

df = df.cumsum()
maxP = list(df.max(axis=0))

# Create subplots for each column with respect to the last column
fig, axs = plt.subplots(nrows=1, ncols=len(df.columns) - 2, figsize=(15, 10))

# Loop through columns and create scatter plots
for i, column in enumerate(df.columns[1:-1]):
    axs[i].plot((0,maxP[-1]),(0,maxP[i+1]),':k',label = '1:1 line')
    axs[i].plot(df['AVG'], df[column],'.-')
    axs[i].set_xlabel('AVG')
    axs[i].set_ylabel(column)
    axs[i].set_title(f'{column} vs. AVG')

plt.tight_layout()
plt.show()
```
